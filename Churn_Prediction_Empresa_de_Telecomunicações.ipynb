# **Churn Prediction para uma empresa de Telecomunicação**

Todas as indústrias sofrem com a rotatividade voluntária - a perda de clientes para alguma outra empresa. A sobrevivência de qualquer empresa baseia-se em sua capacidade de reter clientes, isso porque modelos de negócio baseados na economia da recorrência têm sua receita brutalmente impactada por esses clientes: junto com suas assinaturas, eles levam também a força da previsibilidade de receita do seu negócio.

As empresas normalmente gastam a maior parte de seus esforços e recursos na aquisição de clientes, embora estudos comprovem que o *custo de manter um cliente existente seja de 5 a 25 vezes menor do que adquirir um novo*. De acordo com a pesquisa feita por **Frederick Reichheld da Bain & Company**, aumentar as **taxas de retenção de clientes** em apenas **5% pode aumentar os lucros em 25% a 95%**.

A retenção de clientes é uma medida de quantos de seus clientes continuam a comprar de você ao longo do tempo e, portanto, são leais à sua marca. A rotatividade, às vezes conhecida como atrito ou churn com o cliente, está na extremidade oposta desta afirmação, ou seja, quantos clientes param de comprar de sua empresa.

## **O que é Customer churn?**

**Customer Churn** (também conhecida como Atrito de Clientes ou Rotatividade de Clientes) é uma tendência dos clientes de abandonar uma marca e deixar de ser um cliente pagador de uma determinada empresa. A porcentagem de clientes que deixam de usar os produtos ou serviços de uma empresa durante um determinado período de tempo é chamada de taxa de rotatividade (atrito) de clientes. Uma das maneiras de calcular uma taxa de rotatividade é dividir o número de clientes perdidos durante um determinado intervalo de tempo pelo número de clientes adquiridos e, em seguida, multiplicar esse número por 100 por cento. Por exemplo, se você conquistou 150 clientes e perdeu três no mês passado, sua taxa de rotatividade mensal é de 2%.

Os setores que usam um modelo de negócios baseado em assinatura tradicionalmente se concentram mais na rotatividade do que em outros. Bancos, empresas de telecomunicações, seguradoras, empresas de serviços de energia estão entre os muitos tipos de negócios que costumam usar a análise de churn de clientes e as taxas de rotatividade de clientes como uma de suas principais métricas de negócios.

### **Por quê empresas devem se preocupar com o churn de seus clientes?**

Há muitas coisas que as marcas podem fazer de errado, desde uma integração complicada quando os clientes não recebem informações fáceis de entender sobre o uso do produto e seus recursos até uma comunicação deficiente, por exemplo, a falta de feedback ou atrasos nas respostas às perguntas. Outra situação: clientes antigos podem se sentir desvalorizados porque não recebem tantos bônus quanto os novos.

Em geral, é a experiência geral do cliente que define a percepção da marca e influencia como os clientes reconhecem o valor pelo dinheiro dos produtos ou serviços que usam.

A realidade é que mesmo os clientes fiéis não tolerarão uma marca se tiverem um ou vários problemas com ela. Por exemplo, 59% dos entrevistados nos Estados Unidos da pesquisa da [PricewaterhouseCoopers (PwC)](https://www.pwc.com/us/en/advisory-services/publications/consumer-intelligence-series/pwc-consumer-intelligence-series-customer-experience.pdf#page=8) observaram que vão dizer adeus a uma marca após várias experiências ruins, e 17% deles após apenas uma experiência ruim.

### **Previsão de rotatividade de clientes com Machine Learning**

Como dito anteriormente, o churn pode acontecer devido a muitos motivos diferentes e a análise do churn ajuda a identificar a causa (e o momento) desse churn, abrindo oportunidades para implementar estratégias de retenção eficazes. Podemos citar 6 etapas utilizadas por empresas para garantir a retenção de seus clientes - neste projeto nos concentraremos apenas nas etapas 2 e 3. 

  01. Reúna o comportamento do cliente, transações, dados demográficos e padrões de uso disponíveis
  02. Utilize esses dados para prever os segmentos de clientes que provavelmente se desligarão
  03. Crie um modelo para padronizar a tolerância ao risco do negócio em relação à probabilidade de rotatividade.
  04. Projetar um modelo de intervenção para considerar como o nível de intervenção pode afetar as porcentagens de rotatividade e o valor de vida útil do cliente (CLV)
  05. Implementar experimentação eficaz em vários segmentos de clientes para reduzir a rotatividade e promover a retenção.
  06. Enxágue e repita a partir da Etapa 1 (o gerenciamento da rotatividade cognitiva é um processo contínuo e não um exercício anual).

Um modelo de churn prediction é uma ferramenta de **classificação simples**: ele observa a atividade do usuário no passado e verifique quem está ativo após um determinado tempo e, em seguida, crie um modelo que identifique probabilisticamente as etapas e estágios quando um cliente (ou segmento) está saindo seu serviço ou produto.

Ter um modelo preditivo como este fornece consciência e métricas quantificáveis ​​para combater em seus esforços de retenção. Isso dá as empresas a capacidade de padronizar os hábitos dos clientes que cancelam suas assinaturas e intervir antes que eles tomem essa decisão. Sem essa ferramenta, empresas estariam agindo com base em suposições gerais, não em um modelo baseado em dados que reflete como seus clientes realmente agem.

Sem um forte entendimento de seus clientes e de seus comportamentos, é difícil retê-los, então a primeira etapa na criação deste modelo é entender o comportamento de seu cliente a partir de pontos nos dados recolhidos. Vamos ver que tipo de dados precisamos para avaliar os gatilhos que os levaram a deixar sua empresa.

O objetivo principal deste projeto se baseia em desenvolver a criação modelo de Machine Learning de **Classificação Supervisionado** capaz de prever a rotatividade de clientes com base nos dados disponíveis e assim diminuir os prejuízos destes possíveis atritos e conseqüentemente aumentar o lucro da empresa em questão.

#### **Diferença da Proporção na População de Churn e Não Churn em cada classe das Variáveis Categóricas**

#### Questão de pesquisa

Existe alguma **diferença significativa entre as proporções populacionais** de pessoas que **não efetuam Churn** (não cancelamento de seus serviços com a empresa de telecom em questão) e pessoas que **efetuam Churn** (cancelamento de seus serviços com a empresa de telecom) em cada categoria das variáveis categóricas presentes neste Dataset?

* **Populações**: pessoas que cancelam seus serviços e pessoas que não cancelam seus serviços com a empresa de Telecom

* **Teste bicaudal**: Identificar diferença significativa entre as proporções pessoas que cancelam seus serviços e pessoas que não cancelam seus serviços com a empresa de Telecom

<p align=center><img alt="z-test" src='https://drive.google.com/uc?id=1qLTyHvcONXHPFRzjQ4brelEH5amEychg'></p>

* **Parâmetro de Interesse**: `Proporção Não Churn` - `Proporção Churn`  

* **Hipótese Nula (H0):** `Proporção Não Churn` - `Proporção Churn` = 0 ou `Proporção Não Churn` = `Proporção Churn`

* **Hipótese Alternativa (HA):** `Proporção Não Churn` - `Proporção Não Churn` $\neq$ 0  ou `Proporção Não Churn` $\neq$ `Proporção Não Churn`

* **Premissas para este Teste**
  * Amostras usadas são consideradas *Amostras Aleatórias Simples*
  * As Variáveis Churn e Não Churn são *Independentes* uma da outra
  * Ambas Populações possuem *Distribuições aproximadas Normais* ou o *Tamanho das Amostras* são grandes o suficiente (*Tamnho Proporcional das amostras maior ou igual a 10*) para ambos os grupos Churn e Não Churn

##### **Teste Estatísticos de Diferença Proporcional entre os casos de Churn e não Churn**

```
print('-'*102)
titulo= f'\033[1;30mTeste Hipotético de Proporção : Duas Variáveis (Yes e No - Churn)\033[0;30m'
print(titulo.center(118))

for j in category_cols:
  df_cross = pd.crosstab(index=df_work["Churn"], columns=df_work[j], margins=True)
  df_cross_t = df_cross.T

  # lista tamanho da amostra
  list_sample_size = df_cross_t.values[len(df_cross_t.values)-1]

  # definindo tamanho da amostra para nao churn e churn
  nao_churn = list_sample_size[0]
  churn = list_sample_size[1]

  print('-'*102)
  variavel = f'\033[1;30m{j}\033[0;30m'
  print(variavel.center(118))
  print('-'*102)
  cat = 'Categoria'
  print(f'| \033[1;30m{cat: <25}\033[0;30m', end = ' | ')
  ts = 'Teste Estatístico (Z)'
  print(f" \033[1;30m{ts: >20}\033[0;30m", end = ' | ')
  pv = 'P-Value'
  print(f" \033[1;30m{pv: >20}\033[0;30m", end = ' | ')
  rs = 'Resultado'
  print(f' \033[1;30m{rs: >20}\033[0;30m |')
  print('-'*102)
  # Numero de casos de nao churn e churn para cada classe
  for index,classe in enumerate((df_cross_t.index)[:-1]):
    
    print(f'| {classe: <25}', end= ' | ')
    # Numero de casos de nao churn e churn
    num_caso_no = df_cross_t.values[index][0]
    num_caso_yes = df_cross_t.values[index][1]

    # Estimando proporção da população
    p_no = round((num_caso_no/nao_churn),2)
    #print(f'população no = {p_no}')
    p_yes = round((num_caso_yes/churn),2)
    #print(f'população yes = {p_yes}')

    # Estimando a proporção combinada da população
    phat = (num_caso_no + num_caso_yes) / (nao_churn + churn)
    #print(f'Prop Populacao Estimada = {phat}')

    # Estimando a variancia da proporção combinada da população
    var = phat * (1 - phat)
    #print(f'Variancia da proporção combinada da população = {var}')

    # Estimando o desvio padrao da proporção combinada da população
    se = np.sqrt(var * (1 / nao_churn + 1 / churn))
    #print(f'Desvio padrao da proporção combinada da população = {se}')

    # Teste estatistico e seu p-value
    test_estat = (p_no - p_yes) / se
    pvalue = 2*dist.norm.cdf(-np.abs(test_estat))

    # Imprimindo o Teste Estatistico e o p-value
    #print("Teste Estatístico (Z)")
    print(f' {round(test_estat, 2): >21}', end= ' | ')
    #print("\nP-Value")
    print(f' {round(pvalue, 3): >20}', end= ' | ')
    alpha = 0.05
    if round(pvalue, 3) < alpha:
      frase = 'Rejeitar a H0'
    else:
      frase = 'Aceitar a H0'
    print(f'{frase: >21} |')
    #print()
    print('-'*102)
```

Observando o gráfico acima podemos detectar que existem algumas diferenças visuais de proporção entre casos de **Não Churn** e **Churn** em cada categoria das variáveis presente no dataset.

Efetuamos um Teste Estatístico (T) a fim determinar se a diferença propocional entre os casos acima citados nos mostram uma diferença populacional significativa entre as classes de cada variável dada a proporção amostral existente neste conjunto de dados. 

Essa diferença poprocional em casos não churn e churn pode nos indicar três fatores:

#### **Diferença Média de Pagamento Mensal na População de Churn e Não Churn em cada classe das Variáveis Categóricas**

#### Questão de pesquisa

Existe alguma **diferença significativa entre as médias populacionais** de pagamento mensal (coluna `MonthlyPayment`) entre pessoas que **não efetuam Churn** (não cancelamento de seus serviços com a empresa de telecom em questão) e pessoas que **efetuam Churn** (cancelamento de seus serviços com a empresa de telecom) em cada categoria das variáveis categóricas presentes neste Dataset?

* **Populações**: pessoas que cancelam seus serviços e pessoas que não cancelam seus serviços com a empresa de Telecom

* **Teste bicaudal**: Identificar diferença significativa entre as proporções pessoas que cancelam seus serviços e pessoas que não cancelam seus serviços com a empresa de Telecom

<p align=center><img alt="z-test" src='https://drive.google.com/uc?id=1qLTyHvcONXHPFRzjQ4brelEH5amEychg'></p>
 
* **Parâmetro de Interesse**: (𝜇1 - 𝜇2) na Variável `MonthlyCharges` 

* **Hipótese Nula (H0):** 𝜇1 - 𝜇2 = 0 ou 𝜇1 = 𝜇2

* **Hipótese Alternativa (HA):** 𝜇1 - 𝜇2 $\neq$ 0  ou 𝜇1 $\neq$ 𝜇2

* **Onde**:
  * 𝜇1 = `Média de Pagamento Mensal de clientes que não deram Churn` 
  * 𝜇2 = `Média de Pagamento Mensal que deram Churn`

* **Premissas para este Teste**
  * Amostras usadas são consideradas *Amostras Aleatórias Simples*
  * As Variáveis Churn e Não Churn são *Independentes* uma da outra
  * Ambas Populações possuem *Distribuições aproximadas Normais* ou o *Tamanho das Amostras* são grandes o suficiente (*maior que 200 amostras*) para ambos os grupos Churn e Não Churn

##### **Teste Estatístico da Diferença nas Médias de Pagamento Mensal entre casos de Churn e Não Churn**

```
print('-'*102)
titulo= f'\033[1;30mTeste Hipotético da Diferença de Média quanto ao Monthly Payment : Duas Variáveis (Yes e No - Churn)\033[0;30m'
print(titulo.center(118))

for index, var in enumerate(category_cols):
  print('-'*102)
  variavel = f'\033[1;30m{var}\033[0;30m'
  print(variavel.center(118))
  print('-'*102)
  cat = 'Categoria'
  print(f'| \033[1;30m{cat: <25}\033[0;30m', end = ' | ')
  ts = 'Teste Estatístico (Z)'
  print(f" \033[1;30m{ts: >20}\033[0;30m", end = ' | ')
  pv = 'P-Value'
  print(f" \033[1;30m{pv: >20}\033[0;30m", end = ' | ')
  rs = 'Resultado'
  print(f' \033[1;30m{rs: >20}\033[0;30m |')
  print('-'*102)

  #print(c)
  mean_test = df_work[['Churn', var,'MonthlyCharges']].copy()
  
  # definindo df no chur
  df_no_churn = mean_test.loc[mean_test.Churn == 'No']

  # definindo df chur
  df_churn = mean_test.loc[mean_test.Churn == 'Yes']

  # criando lista com categorias
  list_cat = list(df_churn[var].unique())

  for j in list_cat:
    Ho = 0
  
    #print(j)
    print(f'| {j: <25}', end= ' | ')
    no_month_col = df_no_churn.loc[df_no_churn[var] == j].MonthlyCharges
    yes_month_col = df_churn.loc[df_churn[var] == j].MonthlyCharges

    no_desc = list()
    yes_desc = list()

    # No Churn mean, std, var e count
    no_desc.append(round(no_month_col.mean(),3))
    no_desc.append(round(no_month_col.std(),3))
    no_desc.append(round(no_month_col.var(),3))
    no_desc.append(no_month_col.count())

    # Yes Churn mean, std, var e count
    yes_desc.append(round(yes_month_col.mean(),3))
    yes_desc.append(round(yes_month_col.std(),3))
    yes_desc.append(round(yes_month_col.var(),3))
    yes_desc.append(yes_month_col.count())

    #print(f'no_desc = {no_desc}')
    #print(f'yes_desc = {yes_desc}')
    
    # Pooled approach
    if (no_desc[2]**2) == (yes_desc[2]**2):
      num = (no_desc[0] - yes_desc[0] - Ho)
      den_1 = (np.sqrt((((no_desc[3]-1)*(no_desc[1])**2) + \
                        ((yes_desc[3]-1)*(yes_desc[1])**2))/((no_desc[3] + yes_desc[3] -2))))
      
      den_2 = np.sqrt((1/no_desc[3]) + (1/yes_desc[3]))

      z_test = round(num / (den_1*den_2), 3)

    # Unpooled approach
    else:
      num = (no_desc[0] - yes_desc[0] - Ho)
      den = (np.sqrt((((no_desc[1])**2)/no_desc[3]) + (((yes_desc[1])**2)/yes_desc[3])))
      z_test = round(num/den, 3)

    import scipy.stats
    #print(f'z-test = {z_test}')
    print(f' {z_test: >21}', end= ' | ')
    # Encontrando p-value para teste de cauda única
    p_value = round((scipy.stats.t.sf(abs(z_test), df=[no_desc[3] - yes_desc[3] -2]))[0], 3)*2
    #print(f'p-value = {p_value}')
    print(f' {p_value: >20}', end= ' | ')
    alpha = 0.05

    if p_value < alpha:
      frase = 'Rejeitar a H0'
    else:
      frase = 'Aceitar a H0'
    print(f'{frase: >21} |')
    print('-'*102)
    
```

#### **Chi-Square Teste das Variáveis Categóricas**

Pearson's Chi-Square é um teste estatítstico de hipótese que verifica a independência entre variáveis ​​categóricas. Aqui realizaremos este teste usando o módulo `SciPy` do Python.

Etapas para se realizar o Chi-Sqaure Teste:

**Etapa 1: Definir a hipótese**
  * Um Teste de Hipótese é um método estatístico que avalia duas afirmações (hipótese) e determina qual delas é verdadeira. No teste de hipótese, a afirmação inicial é chamada de Hipótese Nula denotada como `H0` e a segunda que geralmente é complementar à primeira é chamada de Hipótese Alternativa denotada como `H1`.
  Como exemplo temos as hipóteses:
    * `H0`: Os recursos Gênero e Churn são independentes (o que significa que não estão associados).
    * `H1`: Gênero e Churn não são dependentes (o que significa que estão associados).

**Etapa 2: Escolha um nível de significância**
 * Basicamente, o nível de significância denotado por α (alfa) é uma medida de quão certos queremos ter sobre nossos resultados.
 Ou em palavras simples, o nível de significância é usado para determinar se a hipótese nula deve ser rejeitada ou não. Para que a hipótese nula seja rejeitada, o valor p (explicado posteriormente) deve ser menor que o nível de significância.
Valores de α mais baixos são geralmente preferidos, os quais podem estar na faixa de 0,01 a 0,10.
  * Escolhemos α = 0,05

**Etapa 3: Criar tabela de contingência**
  * Uma tabela que mostra a distribuição de frequência de uma variável em linhas e outra em colunas, usada para estudar a correlação entre as duas variáveis, é chamada de *Tabela de Contingência* que pode ser criada o método `crosstab`do Pandas.

**Etapa 4: Calcular a frequência esperada**

Calculamos a contagem de frequência esperada para cada célula. A fórmula para contar a frequência esperada é:

$$E = \frac{(Total da Linha * Total da Coluna)}{Total Geral}\
$$

**Etapa 5: Calcule o valor de Chi-Square ou estatística Chi-Square**
  * A fórmula para calcular o valor Chi-Square ou 𝜒2 é:

\begin{equation}
\chi^2=\Sigma\frac{(O-E)^2}{E} \\
\text{Onde O é o Valor Atual e E é Valor Esperado.}
\end{equation}

Conforme mostrado, 𝜒2 é a soma da diferença quadrática entre as frequências observadas e esperadas dividido pela frequência esperada para todas as células.

**Etapa 6: Calcular Graus de Liberdade**
  * Os graus de liberdade podem ser calculados como:

$${Grau}{ }{Liberdade} = \{({No}{Linhas} - 1) * ({No}{Colunas} - 1)}\$$

**Passo 7: Encontre o valor p**
  * Podemos ver as tabelas de distribuição do Chi-square como [esta](http://kisi.deu.edu.tr/joshua.cowley/Chi-square-table.pdf) para encontrar o valor p usando os valores de Chi-square, graus de liberdade e seu α (alpha = Nível de Significância).

**Passo 8: Decida se deve rejeitar ou manter a hipótese nula**
  * Em palavras simples, se nosso valor de p for menor que o valor de significância alpha (α), rejeitamos a hipótese nula e se nosso valor de p for maior que o valor de significância, não a rejeitaremos.

O teste Chi-Square irá medir a dependência entre as variáveis ​​estocásticas, portanto, podemos **eliminar** as características que têm maior probabilidade de serem **independentes** do valor target usando esta função. É basicamente para avaliar se a diferença entre dois grupos separados (*Variáveis Catgóricas x Churn*) de amostras não negativas é por acaso ou não.

O módulo `SciPy` fará todo cálculo necessário para realizar este teste.

Utilizaremos o processo de `cross validation` do *Scikit-Learn* que dividirá nosso dataset de treino em **dez partes** onde **nove** serão utilizadas para treino e **uma** para teste onde salvaremos todos os resultados e colheremos o resultado médio destes testes, assim poderemos comprar os modelos e escolher os melhores para nosso teste final.

```
# função para retornar os resultados de cada modelo
def results(dicionario, X, y, num_blocos, seed, score):

  dicionario_resultados = {}

  # Testando os modelos do dic_model e computando seus resultados e nomes
  for nome, modelo in dicionario.items():
    # método k-fold
    kfold = StratifiedKFold(n_splits=num_blocos, random_state=seed)
    # método cross validation
    cv_results = cross_val_score(modelo, X, y, cv=kfold, scoring=score)
    # salvando nome e resultados em um dicionario
    dicionario_resultados[nome] = cv_results

  return dicionario_resultados

# funcao para criar o dataframe com resultados médios
def dataframe(nomes, resultados):
  padrao_medias = []
  padrao_std = []

  for n in resultados:
    padrao_medias.append(round(n.mean(), 3))
    padrao_std.append(round(n.std(), 3))

  dicionario = {}
  dicionario['Modelos'] = nomes
  dicionario['Mean'] = padrao_medias
  dicionario['Std'] = padrao_std

  df_modelo_padrao = pd.DataFrame(dicionario, columns=['Modelos','Mean', 'Std']).set_index('Modelos')

  
  return df_modelo_padrao.T

#funcao para criar graficos boxplots
def boxplots(nomes, resultados, figsize=(8, 5)):
  # Criando Dataframe para grafico seaborn
  D = dict()
  for i,n in enumerate(resultados):
    D[list(nomes)[i]] = n

  df_boxplot = pd.DataFrame(D, columns=list(nomes))
  df_melt = pd.melt(df_boxplot.reset_index(), id_vars=['index'], value_vars=df_boxplot.columns)

  # Inicializando a figura e setando seu tamanho
  fig, ax = plt.subplots(figsize = figsize)

  # Plotando grafico
  sns.boxplot(data = df_boxplot, orient='h', whis=[0, 100], width=.6, palette="vlag")
  sns.swarmplot(x="value", y="variable", data=df_melt, color='#7d0013')

  # setando nome dos eixos e padronizando eixos com os graficos anteriores
  #ax.xaxis.grid(True)
  ax.set(ylabel="Modelos")
  ax.set(xlabel="Pontuação")
  ax.set_title(('Comparação Algoritmos\n'))
  sns.despine(offset=10, trim=True)

  plt.tight_layout()
```

```
# dicionario com modelos base
dic_models = {
    'LR' : LogisticRegression(),
    'LDA': LinearDiscriminantAnalysis(),
    'KNN': KNeighborsClassifier(),
    'DT': DecisionTreeClassifier(),
    'NB': GaussianNB(),
    'SVM': SVC(),
    'SGD': SGDClassifier(),
    'AB': AdaBoostClassifier(),
    'GBM': GradientBoostingClassifier(),
    'RF': RandomForestClassifier(),
    'ET': ExtraTreesClassifier(),
    'XGB': XGBClassifier(objective= 'binary:logistic')
}

# Características do Teste options e metrica de avaliação
num_blocos = 10
seed = 42
score = 'recall'

# calculando resultados base com funcao results
dic_resultados_base = results(dic_models, X_train_mod, y_train, num_blocos, seed, score)

print(f'\033[1;30mAvaliação de Recall dos Algoritmos Base de Machine Learning\n\033[0;30m')
# Imprimindo dataframe com resultados médios
dataframe_resultado_base = dataframe(dic_resultados_base.keys(), dic_resultados_base.values())
dataframe_resultado_base
```

### **Feature Engineering**

Feature Engineering se refere ao processo de criação de novas variáveis ​​de entrada a partir dos dados disponíveis. Os novos recursos criados a partir do processo de Feature Engineering são altamente específicos para cada conjunto de dados e tipos de dados.

   * Adicionar uma variável de sinalizador booleano para algum estado.

#### **Discretizando Variáveis Contínuas**

Basicamente uma transformação de discretização transforma variáveis ​​numéricas em valores discretos.

Os valores da variável são agrupados em compartimentos discretos e cada compartimento é atribuído a uma variável correspondente à aquele grupo em questão.

O uso de agrupamento é freqüentemente referido como binning, onde k se refere ao número de grupos aos quais uma variável numérica é mapeada.

O mapeamento fornece uma classificação grupal dos valores numéricos que pode suavizar as relações entre as observações contidas nesta variável numérica em questão. 

A transformação pode ser aplicada a cada variável numérica no conjunto de dados de treinamento e teste e, em seguida, fornecida como entrada para um modelo de Machine Learning para aprender uma tarefa de modelagem preditiva.

Aqui criaremos os seguintes grupo:

  * Variável `tenure`
    * *0 Anos*
    * *1 Ano*
    * *2 Ano*
    * *3 Anos*
    * *4 Anos*
    * *+5 Anos*

```
# criando funcoes para divisao de grupos das variaveis numericas

# mudando a variável tenure para uma coluna categórica
def tenure_split(df) :   
    if df['tenure'] < 12 :
        return '0 Anos'
    elif (df['tenure'] >= 12) & (df['tenure'] < 24 ):
        return '1 Ano'
    elif (df['tenure'] >= 24) & (df['tenure'] < 36) :
        return '2 Anos'
    elif (df['tenure'] >= 36) & (df['tenure'] < 48) :
        return '3 Anos'
    elif (df['tenure'] >= 48) & (df['tenure'] < 60) :
        return '4 Anos'    
    elif (df['tenure'] >= 60) :
        return '+5 Anos'

# criando coluna tenure_grupo no dataframe de treino
X_train_mod_feat_1['tenure_grupo'] = X_train_mod_feat_1.apply(lambda X_train_mod_feat_1:tenure_split(X_train_mod_feat_1), axis = 1)

# criando coluna tenure_grupo no dataframe de teste
X_test_mod_feat_1['tenure_grupo'] = X_test_mod_feat_1.apply(lambda X_test_mod_feat_1:tenure_split(X_test_mod_feat_1), axis = 1)

```

#### **Dimensionamento das Variáveis dos Conjuntos de Treino e Teste**

Muitos algoritmos de Machine Learning possuem melhor desempenho quando as variáveis ​​de entrada numéricas são dimensionadas para um intervalo padrão. Isso inclui algoritmos que usam uma soma ponderada da entrada, como regressão linear, e algoritmos que usam medidas de distância, como KNN. 

As duas técnicas mais populares para dimensionar dados numéricos antes da modelagem são normalização e padronização. A normalização dimensiona cada variável de entrada separadamente para o intervalo 0-1, que é o intervalo para valores de ponto flutuante onde temos mais precisão. A padronização dimensiona cada variável de entrada separadamente subtraindo a média (chamada de centralização) e dividindo pelo desvio padrão para deslocar a distribuição para ter uma média de zero e um desvio padrão de um.

Neste projeto utilizaremos a Padronização utilizando o método `StandardScaler` do *Scikit-Learn*. Nossos dados numéricos ficaram intervalados entre -1 e 1.

Nesta etapa determinaremos quais variáveis mais importantes são as mais importantes para que cada algoritmo obtenha o maior valor de eficácia.

Para isso utilizaremos o método do *Scikit-Learn* `RFECV` que significa **Recursive Feature Elimination with Cross-Validation**, ou seja, Eliminação Recursiva com processo de Validação Cruzada.

A eliminação recursiva de variáveis é basicamente uma seleção retroativa das variáveis preditoras juntamente com o processo de validação cruzada. Essa técnica começa construindo um modelo em todo o conjunto de dados e computando uma pontuação de importância para cada variável preditora. 

Os variáveis preditoras menos importantes são então removidas, o modelo é reconstruído e as pontuações de importância são calculadas novamente. 

Na prática, especificaremos o número de subconjuntos preditores a serem avaliados, bem como o tamanho de cada subconjunto. Portanto, o tamanho do subconjunto é um parâmetro de ajuste para RFECV. O tamanho do subconjunto que otimiza os critérios de desempenho é usado para selecionar os preditores com base nas classificações de importância. O subconjunto ideal é então usado para treinar o modelo final.

Esta técninca não pode ser usada no algoritmo **KNN** não fornece lógica linear para fazer a seleção de recursos e a técninca  `RFECV` só pode ser usado quando o algoritmo possui esta lógica. Com isso, deixaremos o algoritmo **KNN** fora desta análise e daremos proseguimento com os outros.

Abaixo imperentaremos este processo para cada um dos cinco modelos restantes analisados anteriormente.


```
best_models = {
    'GBM': GradientBoostingClassifier(),
    'RF': RandomForestClassifier(),
    'ET': ExtraTreesClassifier(),
    'XGB': XGBClassifier(objective= 'binary:logistic'),
}

# titulo
variavel = f'\033[1;30mEliminação Recursiva de Variáveis\033[0;30m\n'
print(variavel.center(115))

# listas para salvar colunas a eliminar, numero optimal de colunas e pontuacoes obtidas
lista_colunas_elim = []
no_optimal = []
pontuacoes = []

# processo de loop sobre dicionario para aplicacao do rfecv para cada modelo
for nome, modelo in best_models.items():

  print('-'*118)
  variavel = f'\033[1;30m{nome}\033[0;30m'
  print(variavel.center(115))
  print('-'*118)

  # intanciando modelo
  model = modelo
  # instanciando o rfecv
  rfecv = RFECV(estimator=model, step=1, cv=StratifiedKFold(num_blocos), scoring=score)
  # treinando o rfecv
  rfecv.fit(X_train_adasyn_scaled, y_train_adasyn_scaled)

  # enviando as pontuacoes para sua lista criada acima
  pontuacoes.append(rfecv.grid_scores_)
  # enviando as colunas para eliminacao para sua lista criada acima
  lista_colunas_elim.append(list(np.where(rfecv.support_ == False)[0]))
  # enviando os numeros optimos para sua lista criada acima
  no_optimal.append(rfecv.n_features_)

  # printando resultados
  print(f'\033[1;30mNo Optimal:\033[0;30m {rfecv.n_features_: <32}')

  if len(np.where(rfecv.support_ == False)[0]) <= 12:
    print(f'\033[1;30mCol. Eliminação.:\033[0;30m {np.where(rfecv.support_ == False)[0]}')
  else:
    print(f'\033[1;30mCol. Eliminação.:\033[0;30m {np.where(rfecv.support_ == False)[0][:12]}')
    print(f'\033[1;30mCol. Eliminação.:\033[0;30m {np.where(rfecv.support_ == False)[0][12:]}')
  print('-'*118)
```

#### **Tunando Hyperparâmetros**

Os modelos de Machine Learning são parametrizados para que seu comportamento possa ser ajustado para um determinado problema. Os modelos podem ter muitos parâmetros e encontrar a melhor combinação de parâmetros pode ajudar cada algoritmo a melhorar sua performance.

O ajuste do algoritmo é uma etapa final no processo de aprendizado de máquina aplicado antes de finalizar seu modelo. 

O método **Random Grid Search** é uma técnica em que combinações aleatórias dos hiperparâmetros são usadas para encontrar a melhor solução para o modelo construído. Ele tenta combinações aleatórias de uma gama de valores. Para otimizar com pesquisa aleatória, a função é avaliada em algum número de configurações aleatórias no espaço de parâmetro. Utilizaremos como número padrão 20 interações em cada algoritmo. 

Por questões computacionais utilizaremos o método `Random Grid Search` do *Scikit-Learn* a fim de encontrar a melhor combinação de hyperparâmetros para cada modelo acima e comprá-los posteriormente.

#### **Comparando Estatísticamente os resultados dos Modelos Analisados**

ANOVA significa 'Análise de Variância' e é um teste abrangente que testa a diferença geral entre grupos independentes. A *ANOVA One-way*, também conhecida como ANOVA de um fator, é um teste estatítisco usado para avaliar uma diferença estatisticamente significativa entre 3 ou mais grupos. Uma vez que é um teste abrangente, ele testa uma diferença geral, ou seja, pelo menos um dos grupos é estatisticamente significativamente diferente dos outros. No entanto, se a *ANOVA* for significativa, não se pode dizer qual grupo é diferente. Para saber qual grupo é diferente, é necessário realizar comparações planejadas ou post-hoc. Como acontece com todos os testes paramétricos, existem certas condições que precisam ser atendidas para que os resultados do teste sejam considerados confiáveis.

Aqui usaremos o `scipy` a fim de calcular o **F-value** e **P-value** e determinar a existência ou não de uma diferença entre grupos analisados.

***Hipóteses do Teste ANOVA One-way***

**Hipótese nula**: as médias dos grupos são iguais (sem variação nas médias dos grupos)
  * **H0**: μ1 = μ2 =… = μp

**Hipótese alternativa**: pelo menos, a média de um grupo é diferente de outros grupos
* **H1**: Pelo menos uma das μ (médias) é diferente

##### **Teste Estatístico ANOVA One-way**

```
# criando dicionario para criacao dataframe para analise ANOVA
dicio_anova = {}
dicio_anova['GBM'] = resultados[0]
dicio_anova['RF'] = resultados[1]
dicio_anova['ET'] = resultados[2]
dicio_anova['XGB'] = resultados[3]
dicio_anova['KNN'] = resultados[4]


# Criando dataframe para teste ANOVA
df_anova = pd.DataFrame(dicio_anova, columns=dicio_anova.keys())

# stats.f_oneway usam os grupos como entrada e retornam ANOVA F-Value e o valor p-value
fvalue, pvalue = stats.f_oneway(df_anova['GBM'], df_anova['RF'],\
                                df_anova['ET'],df_anova['XGB'], df_anova['KNN'])

print('-'*77)
variavel = f'\033[1;30mTeste ANOVA (One-way) - Alpha = 0.05\033[0;30m'
print(variavel.center(95))
print('-'*77)
tf = 'Teste Estatístico (F)'
print(f'| \033[1;30m{tf: <25}\033[0;30m', end = ' | ')
pv = 'P-Value'
print(f" \033[1;30m{pv: >20}\033[0;30m", end = ' | ')
rs = 'Resultado'
print(f' \033[1;30m{rs: >20}\033[0;30m |')
print('-'*77)

# Imprimindo o Teste Estatistico e o p-value
#print("Teste Estatístico (F)")
print(f'| {round(fvalue, 2): >25}', end= ' | ')
#print("\nP-Value")
print(f' {round(pvalue, 3): >20}', end= ' | ')
alpha = 0.05
if round(pvalue, 3) < alpha:
  frase = 'Rejeitar a H0'
else:
  frase = 'Aceitar a H0'
print(f'{frase: >21} |')
#print()
print('-'*77)
```

Por possuirmos um *P-Value* menor que nosso nível de significância *Alpha* (0.048 x 0.050), podemos dizer que há diferenças estatística significativas entre as médias de **Recall** dos modelos usados no conjunto de treino.

Quais modelos possuem diferença estatística.

##### **Test T entre Par de Modelos**

A partir da análise ANOVA, sabemos que as diferenças entre **Recall** médios nos indiccam que são estatisticamente significativas (diferentes), mas o Teste ANOVA não nos diz quais modelos são significativamente diferentes uns dos outros. 

Para saber os pares de Modelos que possuem diferenças significativas, realizaremos análises de comparação múltipla de pares, utilizando o Teste T entre pares a fim de determinar se dois Modelos possuem **Recall** Médios Estatísticamente Diferentes.

O teste pode ser interpretado com base no valor-p.

* P-Value – Considerando o P-Value entre dois modelos, podemos afirmar que se os modelos possuirem P-Values menores que o nível de significância adotado (P-Value < 0,05) os modelos são significamente diferentes.

```
import itertools
x = df_anova.columns
comb = list(itertools.combinations(x, 2))

print('-'*73)
titulo= f'\033[1;30mTeste Hipotético da Diferença de Recall Médio dos Modelos\033[0;30m'
print(titulo.center(88))

for n in comb:
  valor_t, valor_p = stats.ttest_ind(df_anova[n[0]], df_anova[n[1]])

  print('-'*73)
  variavel = f'\033[1;30m{str(n[0])} e {str(n[1])}\033[0;30m'
  print(variavel.center(89))
  print('-'*73)
  #cat = 'Categoria'
  #print(f'| \033[1;30m{cat: <25}\033[0;30m', end = ' | ')
  ts = 'Teste Estatístico (T)'
  print(f"| \033[1;30m{ts: >20}\033[0;30m", end = ' | ')
  pv = 'P-Value'
  print(f" \033[1;30m{pv: >20}\033[0;30m", end = ' | ')
  rs = 'Resultado'
  print(f' \033[1;30m{rs: >20}\033[0;30m |')
  print('-'*72)

  #print(f'z-test = {z_test}')
  print(f'| {round(valor_t,2): >21}', end= ' | ')
  # Encontrando p-value para teste de cauda única
 
  #print(f'p-value = {p_value}')
  print(f' {round(valor_p,2): >20}', end= ' | ')
  alpha = 0.05

  if round(valor_p,2) < alpha:
    frase = 'Rejeitar a H0'
  else:
    frase = 'Aceitar a H0'
  print(f'{frase: >21} |')
  print('-'*73)
```

Os Modelos que possuem Recall Médio Estatísticamente Diferentes são:

* **GBM** / **XGB** e **ET**
* **GBM** / **XGB** e **KNN**

Os demais não possuem diferença estatística significante.

##### **Gráfico de Curva ROC e AUC**

A curva **AUC-ROC** é uma métrica de desempenho para problemas de classificação binária. **ROC** é uma curva de probabilidade e **AUC** representa o grau ou medida de separabilidade. Ele informa o quanto o modelo é capaz de distinguir entre as classes. Quanto maior a AUC, melhor será o modelo em prever **Não Churns** como **Não Churns** e **Churns** como **Churns**.

```
from sklearn.metrics import roc_curve, roc_auc_score

# Instanciando modelos e criando uma lista

# Definindo resultados em um dataframe
result_table = pd.DataFrame(columns=['modelos', 'fpr','tpr','auc'])

# Treinando modelos e salvando resultados
count = 0
for nome, modelo in best_algo.items():
  # treinando modelo
  model = modelo.fit(lista_x_train[count], y_train_adasyn_scaled)
  # salvando probabilidades 
  yproba = model.predict_proba(lista_x_test[count])[::,1]

  # salvando taxas de falso e verdadeiro positivos 
  fpr, tpr, _ = roc_curve(y_test,  yproba)
  # salvando o valor da area embaixo da curva
  auc = roc_auc_score(y_test, yproba)
  # tabela de resultados
  result_table = result_table.append({'modelos':modelo.__class__.__name__,
                                      'fpr':fpr, 
                                      'tpr':tpr, 
                                      'auc':auc}, ignore_index=True)
  
  count += 1

# Setando nome dos modelos como index da tabela
result_table.set_index('modelos', inplace=True)

fig = plt.figure(figsize=(8,6))

for i in result_table.index:
    plt.plot(result_table.loc[i]['fpr'], 
             result_table.loc[i]['tpr'], 
             label="{}, AUC={:.3f}".format(i, result_table.loc[i]['auc']))
    
plt.plot([0,1], [0,1], color='black', linestyle='--')

plt.xticks(np.arange(0.0, 1.1, step=0.1))
plt.xlabel("Taxa Falso Positivo", fontsize=13)

plt.yticks(np.arange(0.0, 1.1, step=0.1))
plt.ylabel("Taxa Verdadeiro Positivo", fontsize=13)

plt.title('Análise Curva ROC', fontsize=13)
plt.legend(prop={'size':12}, loc='lower right')

sns.despine(offset=0, trim=False)

plt.tight_layout()
```
