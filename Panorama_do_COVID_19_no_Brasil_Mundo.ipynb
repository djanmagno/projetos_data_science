### O que é COVID-19? 

Os coronavírus são uma grande família de vírus comuns em muitas espécies diferentes de animais, incluindo camelos, gado, gatos e morcegos. Raramente, os coronavírus que infectam animais podem infectar pessoas, como exemplo do MERS-CoV e SARS-CoV. Recentemente, em dezembro de 2019, houve a transmissão de um novo coronavírus (SARS-CoV-2), o qual foi identificado em Wuhan na China e causou a COVID-19, sendo em seguida disseminada e transmitida pessoa a pessoa.

A COVID-19 é uma doença causada pelo coronavírus, denominado SARS-CoV-2, que apresenta um espectro clínico variando de infecções assintomáticas a quadros graves. De acordo com a Organização Mundial de Saúde, a maioria (cerca de 80%) dos pacientes com COVID-19 podem ser assintomáticos ou oligossintomáticos (poucos sintomas), e aproximadamente 20% dos casos detectados requer atendimento hospitalar por apresentarem dificuldade respiratória, dos quais aproximadamente 5% podem necessitar de suporte ventilatório.

##### **Avaliando a Capacidade dos 5 Países com mais Mortes no Combate contra o COVID-19**

Para esta análise utilizarei 4 variáveis que com os quias criarei uma espécie de pontuação baseada no maior valor declarado entre os países relacionados no Dataframe (com exceção dos valores Mundias e de Continentes).

No meu entendimento, as colunas selecionadas mostram um forte indicativo de como cada país se portou vem se portando em relação ao COVID-19. As colunas são:

- `stringency_index` = Este valor é o Índice de Restrição de Resposta do Governo, mede o entendimento do governo quanto restrições impostas a seus cidadãos em relação ao covid-19.
- `total_tests_per_thousand` = Total de testes para COVID-19 por 1.000 pessoas feitos por cada país.
- `total_vaccinations_per_hundred` = Número total de doses de vacinação COVID-19 administradas para 100 pessoas na população total em cada país
- `hospital_beds_per_thousand` = Leitos hospitalares por 1.000 pessoas fornecidos em cada país.

As etapas para montagem desta análise são:

1. Separar todos os países e colunas para análise.

2. Logo em seguida preencherei os valores ausentes `NaN` com o valor uma vez que entendo que estes valores não foram fornecidos pelos países logo estão ausentes.

3. Utilizando o método `groupby`, agruparei todos estes valores e recolherei a **média** de cada variável a fim de ter valores que representam o que foi feito durante todo o período analisado até então.

4. Aqui passarei todos os valores para uma escala entre 0 e 1 com o método `minmax_scaling` que cria essa escala automaticmente. O valor 1 representa o valor máximo encontrado dentre todos os países e 0 corresponde ao menor valor encontrado dentre os países.

5. Em seguida criarei um `Polar Graph` com `Matplotlib` para representar todas essas "notas" de cada variável escolhida para os 5 países com mais mortes em nosso DataFrame.

Essa análise nos mostrará a capacidade de cada um destes 5 países de lidar com o COVID-19 em relação aos melhores valores fornecidos em nosso dataset.



```
#@title
# Lista com dados que não fazem parte da análise
lista_exc = ['Africa', 'Asia', 'Europe', 'Oceania', 'North America','South America','World', 'European Union']

# Separando as variáveis e países a serem analisados e preenchendo os valores NaN com 0
df_capacidade_paises = df_work.loc[(df_work.location.isin(lista_exc) == False)].copy()
df_capacidade_paises.fillna(0, inplace=True)

# Agrupando valores conforme localizacao e calculando a mediana dos valores
df_capacidade_paises_agrup = df_capacidade_paises[['location','stringency_index', 'total_tests_per_thousand',
         'total_vaccinations_per_hundred', 'hospital_beds_per_thousand' ]].groupby('location').mean()

# Criando um novo Dataframe com valores já escalados
df_capacidade_paises_agrup_escal = minmax_scaling(df_capacidade_paises_agrup, columns=['stringency_index', 'total_tests_per_thousand',
                                                                                       'total_vaccinations_per_hundred', 'hospital_beds_per_thousand'])

# Resetando index para poder separar países da análise
df_capacidade_paises_agrup_escal.reset_index(inplace=True)

# Lista com países da análise
lista = ['United States','Brazil','India','United Kingdom','Mexico']

# Dataframe final
df_capac_final = df_capacidade_paises_agrup_escal.loc[df_capacidade_paises_agrup_escal.
                                                      location.isin(lista)].round(3)
df_capac_final = df_capac_final[['location','stringency_index','hospital_beds_per_thousand',
                'total_tests_per_thousand','total_vaccinations_per_hundred']]
df_capac_final.set_index('location',inplace=True)
df_capac_final[['stringency_index','hospital_beds_per_thousand','total_tests_per_thousand',
                'total_vaccinations_per_hundred']] = df_capac_final[['stringency_index','hospital_beds_per_thousand',
                'total_tests_per_thousand','total_vaccinations_per_hundred']]*10
df_capac_final
```

    - `cardiovasc_death_rate`
    - `diabetes_prevalence`
    - `female_smokers`
    - `male_smokers`

Ao correlacionar duas variáveis, buscamos entender a relação entre elas seja ela positiva ou negativa. No contexto deste caso estamos buscando entender a relação de semelhança das demais variáveis com os números mais recentes de todos os países na coluna `total_deaths` do nosso Dataset.

Esta correlação pode ser mensurada através da função de correlação presente em Python e com ela podemos avaliar a intensidade destas correlações através de um gráfico. Para isso vamos fazer:
  * Criar uma Matriz de correlação com os dados númericos de interesse
  * Gerar um *Heatmap* a partir dessa matriz, usando a biblioteca `Seaborn`

```
#@title
# Criando uma Lista com os campos a serem analisados
listaA = ['total_deaths', 'stringency_index','population_density', 
 'aged_65_older','aged_70_older','gdp_per_capita',
 'cardiovasc_death_rate', 'diabetes_prevalence', 'female_smokers',
 'male_smokers']

# Criando dataframe para análise separando somente os países, ou seja, retirando todos os continentes e dados mundiais
df_corr = df_work.loc[(df_work.date == '2021-02-22') & (df_work.location.isin(lista_exc) == False)].copy()

# criar uma matriz de correlação onde 'lista' equivale as variáveis númericas de nosso interesse
correlacao = df_corr[listaA].corr()

# mostrar a matriz de correlação
display(correlacao)
```

```
#@title
# Criando Gráfico de Correlação
plt.figure(figsize=(40,20))
sns.heatmap(data=correlacao, annot=True);
```

###### **Calculando o Número de Correlações Positivas das variáveis selecionadas com a coluna do número Total de Mortes**

```
#@title
# Numero de variáveis na matriz correlacao
total_matriz_variaveis = correlacao.shape[0]

# Numero de variaveis que possuem correlacao postiva com Número Total de Mortes
corr_pos_mortes = correlacao['total_deaths'].loc[(correlacao.total_deaths > 0) \
                                                & (correlacao.total_deaths != 1)].count()

print(f'\n\033[1;30mQuantidade de variáveis com correlação postiva com coluna Total Deaths\033[0;30m\n')
print(f'Das \033[1;32m{total_matriz_variaveis - 1}\033[0;30m variáveis analisadas em nossa matriz de correlação, \033[1;32m{corr_pos_mortes}\033[0;30m \
possuem correlação positiva com o número Total de Mortes, \no que representa cerca de \033[1;31m{((corr_pos_mortes)/(total_matriz_variaveis -1))*100:.2f}%\033[0;30m das variáveis.')
```

###### **Identificando quais as variáveis negativas da nossa Matriz de Correlação**

```
#@title
# Numero de variaveis que possuem correlacao negativa com Número Total de Mortes                                               
var_neg_mortes = list(correlacao['total_deaths'].loc[(correlacao.total_deaths < 0) \
                                                & (correlacao.total_deaths != 1)].index)
print(f'As variáveis \033[1;30m{var_neg_mortes}\033[0;30m possuem correlação negativa \ncom a variável Total Deaths.')
```

Podemos observar acima que as colunas `cardiovasc_death_rate` e `male_smokers` possuem correlação negativa com o número total de mortes o que causa um pequeno estranhamento já que por se tratar de uma doença cardio-respiratória o COVID-19 deveria possuir uma correlação postiva com estas variáveis. O que podemos supor é que as pessoas neste grupos de risco podem ter tido um cuidado muito grande respeitando bastante as medidas protetivas quanto ao COVID-19.

O fato do campo `population_density` possuir uma correlação negativa com o número total de mortes é algo até que natural já que quanto menos concentrada uma população menores são as chances de transmissão do vírus da COVID-19.

###### **Coeficiente de Determinação (rˆ2) das variáveis com correlação positiva em relação ao Número Total de Mortes**

Após obter dados relacionados entre as variáveis, vamos analisar mais a fundo as 7 variáveis que possuem correlações positivas com o Número Total de Mortes.

Correlação, **Pearson's r** ou somente **r** serve para quantificar a relação entre duas variáveis.

Podemos verificar acima que os campos númericos em sua grande maioria apresentam uma correlação positiva com o campo que queremos analisar.

Como correlação não indica casualidade, vamos determinar o **Coeficiente de Determinação (rˆ2)** que é a porcentagem da variação em y (Total de Mortes) explicada pela variação em x em um gráfico cartesiano que possui duas variáveis.

Na nossa análise vamos verificar o gráfico de dispersão das sete variáveis em relação a variável `total_deaths` e depois determinar o coeficiente de determinação de cada uma delas.

```
#@title
# Separando os 10 casos com maior correlação postiva com coluna Total Deaths

print(f'\n\033[1;30mVariáveis com correlação postiva com coluna Total Death\033[0;30m\n')
correlacao['total_deaths'].loc[(correlacao.total_deaths > 0) \
                                                & (correlacao.total_deaths != 1)].round(2).sort_values(ascending=False)
```

```
#@title
# Criando gráficos de disperção para as 6 variáveis
fig = plt.figure(figsize = (20, 10))

ax1 = fig.add_subplot(3,3,1)
sns.regplot(x=df_corr['stringency_index'], y=df_corr['total_deaths'], color="b", ax = ax1);
plt.xlabel('Index de Restrição')

ax2 = fig.add_subplot(3,3,2)
sns.regplot(x=df_corr['aged_70_older'], y=df_corr['total_deaths'], color="r", ax = ax2);
plt.xlabel('Parcela da Pop. com Idade > 70 Anos')

ax3 = fig.add_subplot(3,3,3)
sns.regplot(x=df_corr['aged_65_older'], y=df_corr['total_deaths'], color="g", ax = ax3);
plt.xlabel('Parcela da Pop. com Idade > 65 Anos')

ax4 = fig.add_subplot(3,3,4)
sns.regplot(x=df_corr['gdp_per_capita'], y=df_corr['total_deaths'], color="y", ax = ax4);
plt.xlabel('PIB dos países')

ax5 = fig.add_subplot(3,3,5)
sns.regplot(x=df_corr['female_smokers'], y=df_corr['total_deaths'], color="purple", ax = ax5);
plt.xlabel('Parcela da Pop. Fumante Mulher')

ax6 = fig.add_subplot(3,3,6)
sns.regplot(x=df_corr['diabetes_prevalence'], y=df_corr['total_deaths'], color="pink", ax = ax6);
plt.xlabel('Parcela da Pop. com Diabetes')

plt.tight_layout()
```

Acima podemos vericar graficamente a correlação postiva em relação das variáveis com o número Total de Mortes.

Vamos verificar agora o valor do Coeficiente de Determinação de cada uma destas variáveis.

```
#@title
# Determinando o Coeficiente de Determinação de cada uma das 6 variáveis presentes nos gráficos acima
coef_det = (correlacao['total_deaths'].loc[(correlacao.total_deaths > 0) \
                                                & (correlacao.total_deaths != 1)]**2)\
.sort_values(ascending=False)

print(f'\n\033[1;30mVariáveis e os seis maiores Coeficiente de Determinação em relação a variável Total de Deaths\033[0;30m\n')
(coef_det*100).round(2)
```

Analisando o **Coeficiente de Determinação rˆ2** de cada uma das variáveis podemos inferir que:
  * 7.62% do número Total de Mortes podem ser explicado pela coluna `stringency_index`. Isso nos indica que estes casos podem ter alguma relação com não respeito as medidas de contingência imposta pelos governos.
  * 3,63% das mortes podem ser relacionadas a parte da população com mais de 65 anos.
  * 3,55% das mortes podem ser explicadas a parte da população com mais de 70 anos.
  * 2,27% das mortes podem ser associadas ao PIB interno de cada país.
  * 1,99% das mortes podem ter afetado Mulheres fumantes.
  * 0,32% das mortes podem ter alguma relação com pessoas que possuem Diabetes.

#### **Verificando quanto da População Brasileira já foi infectada e a taxa de mortalidade do COVID-19 no Brasil**

Vamos utilizar a biblioteca `pywaffle` novamente para criar um gráfico de waffle da população brasileira que já foi infectada pelo COVID-19 e ver a taxa de mortalidade no país.

```
#@title
# Criando dataframe para calcular os valores utilizados no gráfico waffle

waf_br = df_brasil[['total_cases','total_deaths','population']].copy().fillna(0)
waf_br.population = waf_br.population.astype('int64')
waf_br.total_cases = waf_br.total_cases.astype('int64')
waf_br.total_deaths = waf_br.total_deaths.astype('int64')
waf_br = waf_br.loc[waf_br.index == '2021-02-22']
waf_br['nao_infect'] = waf_br.population - waf_br.total_cases
waf_br['infec_nao_mortos'] = waf_br.total_cases - waf_br.total_deaths

waf_br[['total_cases','nao_infect','total_deaths','infec_nao_mortos','population']]
```

```
#@title
# plotar gráfico de waffle Proporção de Infectados x Não Infectados no Brasil

valores={'Infectados': waf_br.total_cases.values[0], 'Não Infectados': waf_br.nao_infect.values[0]}
fig = plt.figure(
                FigureClass=Waffle,
                rows=5,
                columns=20,
                values=valores,
                colors=('red', 'lightgreen'),
                legend={'loc': 'upper left',
                        'bbox_to_anchor': (1.1, 1)},
                icons='child', icon_size=16, icon_legend=True, figsize=(10,5))

fig.set_tight_layout(False)
plt.title('Proporção de Infectados x Não Infectados no Brasil\n')

# plotar gráfico de waffle da taxa de Mortalidade no Brasil

valores={'Mortos': waf_br.total_deaths.values[0], 'Sobreviventes': waf_br.infec_nao_mortos.values[0]}
fig = plt.figure(
                FigureClass=Waffle,
                rows=5,
                columns=20,
                values=valores,
                colors=('red', 'lightgreen'),
                legend={'loc': 'upper left',
                        'bbox_to_anchor': (1.1, 1)},
                icons='child', icon_size=16, icon_legend=True, figsize=(10,5))

fig.set_tight_layout(False)
plt.title('\nTaxa de Mortalidade COVID-19 no Brasil\n')

plt.show()
```

Através dos gráficos acima, podemos inferir que aproximadamente **5%** da população brasileira foi infectada pelo Vírus.
