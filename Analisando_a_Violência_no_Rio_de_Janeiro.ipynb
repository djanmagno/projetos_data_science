# Rio de Janeiro e a Violência

Apesar de possuir excelentes atributos e ser uma das melhores cidade brasileira para ser viver, o Rio de Janeiro possui um grande problema, a **violência**.

A violência, principalmente no Rio de Janeiro, está relacionada a diversos fatores como:
  * Tráfico de drogas e suas Facções Criminosas
  * Corrupção
  * Políticas Publicas Defeituosas
  * Falta de Investimento em educação, saneamento, transporte público, saúde e outros que assolam a vida dos cariocas

Sendo o Rio de Janeiro a segunda maior cidade do país, com 6,2 milhões de habitantes, funciona como uma espécie de "vitrine" do país para o mundo por ser o principal destino de turistas estrangeiros. A violência urbana no município é associada ao tráfico de drogas e suas facções. De acordo com dados de 2008 do Ministério da Justiça, o Rio de Janeiro é o quinto estado brasileiro com maior taxa de homicídios, como também possui, de longe, a polícia que mais mata. Em **2008**, foram **1.137** mortos em confrontos com a polícia, taxa sete vezes maior que qualquer outra região do país. Os gastos totais em segurança somam **12%** do orçamento do Estado.

Mas o tráfico, por si só, não justifica o alto índice de criminalidade. Praticamente todas as grandes metrópoles do mundo possuem comércio ilegal de drogas. Estima-se que o mercado consumidor de cocaína em Nova York, por exemplo, seja duas vezes maior que o Rio. A diferença é que nem a cidade americana nem outras europeias testemunham quase que diariamente cenas de guerra nas ruas, como acontece no Rio. Como explicar isso? A resposta é que, no Rio, facções armadas travam lutas pelo controle de territórios, favorecidas por uma rede de corrupção e pelo descaso histórico do poder público em relação às favelas nos morros cariocas. Para que a situação chegasse a esse ponto, cinco fatores foram decisivos: o aumento da oferta e demanda no mercado da cocaína, o surgimento das facções criminosas durante o regime militar (1964-1985), o crescimento das favelas, a corrupção de setores públicos e privados e medidas repressivas e eleitoreiras sem efeito prático.

As favelas dos morros cariocas desempenharam uma função estratégica para os traficantes. Elas cresceram durante os anos 1960, quando o Rio viveu um processo de rápida urbanização e migração, sem que houvesse um planejamento econômico para atender a população. Na ausência do Estado, os donos do morro - como são conhecidos os traficantes responsáveis pela distribuição de drogas no varejo - se tornaram as figuras mais importantes dentro das favelas. Em troca do silêncio dos moradores, os traficantes mantém a ordem e praticam o assistencialismo, distribuindo produtos como remédios e cestas básicas, além de promoverem festas e bailes funks. A carência ainda forneceu às quadrilhas mão mão de obra barata para o negócio ilegal. Para milhares de jovens, sem oportunidades de estudo ou emprego, o tráfico se tornou a única via de acesso a bens de consumo e até mesmo de sustento para suas famílias. Outros fatores, como o glamour do banditismo e o vício em drogas, também acabaram envolvendo jovens de classe média e alta no crime organizado.

Os maiores líderes da facção, porém, estão em presídios de segurança máxima, de onde ainda controlam o tráfico e outras atividades criminosas, por meio de aparelhos celulares e visitas de parentes e advogados. Por esta razão, outro componente importante na engrenagem do tráfico é a corrupção de órgãos do governo e instituições privadas. O dinheiro do tráfico financia desde a proteção de policiais e conivência de agentes penitenciários até a compra de sentenças de juízes.

Antes concentrada, somente na capital, o município / estado do Rio de Janeiro tiveram um marco em sua história. No ano de **2009**, o governo do Rio lançou a Unidade de Polícia Pacificadora (UPP), a partir da experiência piloto no morro de Santa Marta, em Botafogo. Os policiais, ao invés de invadirem periodicamente, ficavam dentro da comunidade, com a intenção de retomar o controle do território e evitar os confrontos armados. O que foi criado para ser um dos grandes projetos de segurança pública acabou se tornando um dos maiores fracassos dentro das comunidades cariocas.

Além de ter levado a violência para municípios do interior Rio de Janeiro, facções "abandonando" o morro e a polícia (milícia) assumindo em seu lugar. A política das UPPs não funcionam de maneira homogênea e a presença dos policiais nas nas UPPs se tornou um fator de conflito.

Um dos setores mais impactados pela violência é o do Turismo. Em 2007, segundo a Confederação de Nacional de Comércio de Bens, Serviços e Turismo (CNC), o turismo do Rio de Janeiro registrou uma perda de R$ 657 milhões entre janeiro e agosto por conta da criminalidade. O montante perdido por conta da violência equivale ao faturamento de 8,9 dias do setor no estado.

O segmento de bares e restaurantes foi o que mais sofreu com o prejuízo tendo uma redução de R$ 332,1 milhões em seu faturamento. Também registraram perda os segmentos de transportes, agências de viagem e locadoras de veículos; hotéis, pousadas e similares; e atividades culturais e de lazer. No geral naquele ano, a violência foi responsável por 29% da perda total de faturamento do setor.

A criminalidade no Rio de Janeiro, que, conforme visto, possui raízes históricas e atravessa sucessivas administrações. A solução para estes problemas não é fácil ou rápida, como também não existe nenhuma fórmula mágica que possa descoberta. Enquanto isso Rio de Janeiro vai sobrevivendo e mantendo seu status de "Cidade Maravilha", que entretando, por trás das cortinas, esconde muito a ser resolvido.

Vamos preenchê-los com valores NaN's que por definição sãovalores 'Not a Number'.
  1. Criar uma cópia do Dataframe para manipulações sem comprometer o Dataframe original
  2. Criar uma função para substituir os espaços em brancos
  3. Aplicar a função criada para ter um dataframe sem espaços em branco
  4. Verificar novamente as 5 primeiras linhas do Dataframe

```
# Criando uma cópia do Dataframe (df_work) para mudanças
df_work = df.copy()

# função para substituir ' ' por valores NaN
def replace_espace(df):
  colunas = df.columns.values

  for coluna in colunas:
    df[coluna].replace(' ', np.nan, inplace = True)
  return df  

# Aplicando função replace_espace e vendo as 5 primeiras colunas
replace_espace(df_work).head()
```

Agora podemos verificar que os espaços em branco ' ', foram substituídos por NaN.

```
# Quantidade de valores ausentes ordem decrescente em relação suas respectivas variáveis
print(f'\033[1;30mPorcentagem de Dados Ausentes por Coluna\033[0;30m\n')
ausentes_por_coluna = (df_work.isnull().sum() / df_work.shape[0]).sort_values(ascending=False)[0:31]
print(round(100 * ausentes_por_coluna, 2))

print(f'\n\033[1;30mPorcentagem de Dados Ausentes em todo Dataset\033[0;30m\n')
total_celulas = np.product(df_work.shape)
total_ausentes = df_work.isnull().sum().sum()
percent_ausentes = round((total_ausentes/total_celulas) * 100, 2)
print(f'{percent_ausentes} %')
```

Para pensar em uma substituição dos valores NaN presentes no dataset, ferei suposições baseadas na data em os registros começaram a fim de dar proseguimento na substituição dos mesmo.

Dataset:

  1. Os dados presentes começam no ano de 1991 e claro que o crescimento dos casos / vítimas é um fator que está diretamente ligado ao crescimento populacional no Estado do Rio de Janeiro, ou seja, a medida que a população cresce, teoricamente teremos um crescimente no número de casos / vítimas em cada variável (crime) presentes no dataset.
  2. Com coluna `roubo_celular` e `furto_celular` é um caso mais complicado de ser analisado, ele só começou a ser registrado apartir do ano 2000. Juntamente quando tivemos um 'boom'no uso de smartfones.
  3. Os outros crimes podem ser considerados que também poderiam ocorrer em anos que não haviam registros dos mesmo no conjunto de dados.

Suposições para fins analíticos:

  * Como os dados se iniciam no ano de 1991 e a coluna `roubo_celular` se inicia no ano 2000. Vou assumir que antes deste ano poucas pessoas possuiam um celular e consequentemente seria mais difícil de ocorrer um delito de tal tipo. **Considerarei as valores NaN's antes de 2000 iguais a 0**
  * Já em relação aos outros delitos, levarei em conta que os mesmo poderiam também acontecer independente do ano inicial de seu registro. E para substituir os valores NaN's nele usarei valores iguais a mediana uma vez que ela apresenta uma melhor representação da medida central de uma população, independente do seu crescimento. **Substituirei os NaN's pelo vamor da mediana**

Vamos utilizar as suposições acima e reaviliar a distribuição / resumo estatístico do novo dataframe.

```
# Criando uma cópia do df_work para termos uma versão final para trabalho
df_final = df_work.copy()

# Substituindo valores NaN's na coluna roubo_celular e furto_celular
df_final.furto_celular.fillna(0, inplace=True)
df_final.roubo_celular.fillna(0, inplace=True)

# Criando uma Função para substituir os valores NaN's pela mediana no restante das colunas e vamos verificar
# as 5 primeiras linhas
def fill_nan(df):
  colunas = df.columns.values

  for coluna in colunas:
    df[coluna].fillna(df[coluna].median(), inplace=True)
  return df.head()

# aplicando a função
fill_nan(df_final)
```

Posso considerar que a primeira falha observada, olhando diretamente para o dataframe. Muitas das variáveis só começaram a ser registradas nos anos subsequentes ao ano mínimo presente neste dataset (1991), o que prejudica um pouco questão de análises dos dados em geral.

Outro fator que me fez questionar um pouco a qualidade destes dados, foi a presença de quedas abruptas no total de casos principalmente próximo ao ano de 2017. Isso seria um motivo para uma investigação mais profunda juntamente com o ISP para ter um entendimento melhor sobre este comportamento.
