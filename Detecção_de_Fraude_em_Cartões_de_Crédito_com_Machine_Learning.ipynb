## **Machine Learning e Detecção de Fraude**

A maioria das soluções de detecção de fraude combinam uma variedade de componentes de dados para formar uma visão conectada de pagamentos genuínos (Sem Fraude) e fraudulentos a fim de decidir sobre a probabilidade de uma transação ser fraudulenta. Cada vez mais, essa visão deve considerar o endereço IP, geolocalização, identificação do dispositivo, dados “BIN”, latitude / longitude global, padrões históricos de transação e as informações reais da transação. Na prática, isso significa a implementação de respostas baseadas em análises que usam dados internos e externos para aplicar um conjunto de regras de negócios ou algoritmos analíticos com objetivo de detectar tais fraudes. As organizações líderes mundias, estão usando Machine Learning e se afastando das regras de negócios para melhorar a detecção de fraudes e reduzir o atrito com seus cliente.

As Instituições avaliam o risco de fraude no momento da autorização do pagamento, normalmente após a autenticação e validação dos fundos disponíveis, com os comerciantes solicitando após o emissor ter autorizado o pagamento.

Existem diversos métodos que vêem sendo usados na detecção de fraude, cada um deles tenta aumentar a taxa de detecção enquanto mantém a taxa de alarmes falsos em seu mínimo. Dentre tais métodos, podemos citar os de classificação como:

* *Algorítmo Naive Bayes*
* *Algorítmo de Regressão Logística*
* *Algorítmo de Árvore de Decisão*
* *Algorítmo Random Forest*
* *etc ...*

Estes métodos podem ser divididos em duas grandes categorias: Supervisionado e Não Supervisionado. Na *Detecção Supervisionada de Fraudes*, os modelos são estimados com base nas amostras de transações fraudulentas e legítimas, a fim de classificar uma nova transação como fraudulenta ou legítima. Enquanto na *Detecção de Fraude Não-Supervisionada*, outliers ou transações incomuns são identificadas como potenciais transações fraudulentas. ambos os métodos de detecção de fraude prevêem a probabilidade de fraude em uma determinada transação de cartão de crédito.

## **Desafios na Detecção de Transações Fraudulentas**

Os principais desafios envolvidos na detecção de fraude de cartão de crédito são:

  1. Dados enormes são processados ​​todos os dias e a construção do modelo deve ser rápida o suficiente para responder ao golpe a tempo.
  2. Dados desequilibrados, ou seja, a maioria das transações (99,8%) não são fraudulentas, o que torna muito difícil detectar as fraudulentas
  3. Disponibilidade de dados, pois os dados são em sua maioria privados.
  4. Dados classificados incorretamente podem ser outro grande problema, pois nem todas as transações fraudulentas são detectadas e relatadas.
  5. Técnicas adaptativas usadas contra o modelo pelos golpistas.

Como enfrentar esses desafios?

  1. O modelo utilizado deve ser simples e rápido o suficiente para detectar a anomalia e classificá-la como uma transação fraudulenta o mais rápido possível.
  2. O desequilíbrio pode ser tratado apropriadamente usando alguns métodos dos quais falaremos mais a frente
  3. Para proteger a privacidade do usuário, a dimensionalidade dos dados pode ser reduzida.
  4. Uma fonte mais confiável deve ser obtida, que verifique os dados, pelo menos para treinar o modelo.
  5. Podemos tornar o modelo simples e interpretável para que, quando o golpista se adaptar a ele com apenas alguns ajustes, possamos ter um novo modelo instalado e funcionando para implantar.

O objetivo deste projeto se baseia em criar um algoritmo de Machine Learning que visa inibir ou evitar transações fraudulentas em cartões de crédito, de forma que os clientes e instituições financeiras não sejam enganados e burlados por fraudadores o que geraria uma economia de Milhões de Dólares / Reais neste setor.

#### **Identificando o quanto as Transações Fraudulentas representam em um contexto geral**

Sabendo que nosso consjunto de dados apresenta transações de cartões de créditos em dois dias seguidos, ou seja, nossa coluna `Time` mostrada em segundos nos mostra o segundo exato das transações (Normais ou Fraudalentas). Com essa informação, vamos determinar, em cada dia, quantas operações fraudulentas tivemos em nosso Dataset mostrando tais informações em uma tabela.

Com estes dados vamos determinar o quanto representaria, de maneira generalizada, o impacto de tais transações em um período de 365 dias (1 ano).

```
# Calculando Quantidade e Montante Totais das transaçoes fraudulentas 
d1_d2f = list(df_work.loc[(df_work.Class == 1), 
                  ['Time', 'Amount']].groupby('Time').agg(['count', 'sum']).sum())

# Calculando Quantidade e Montante do Primeiro Dia das transaçoes fraudulentas
d1f = list(df_work.loc[(df_work.Class == 1) & (df_work.Time < 86400), 
                  ['Time', 'Amount']].groupby('Time').agg(['count', 'sum']).sum())

# Calculando Quantidade e Montante do Segundo Dia das transaçoes fraudulentas
d2f = list(df_work.loc[(df_work.Class == 1) & (df_work.Time >= 86400),
                  ['Time', 'Amount']].groupby('Time').agg(['count', 'sum']).sum())

# Calculando Quantidade e Montante Totais das transaçoes fraudulentas 
d1_d2n = list(df_work.loc[(df_work.Class == 0), 
                  ['Time', 'Amount']].groupby('Time').agg(['count', 'sum']).sum())

# Calculando Quantidade e Montante do Primeiro Dia das transaçoes fraudulentas
d1n = list(df_work.loc[(df_work.Class == 0) & (df_work.Time < 86400), 
                  ['Time', 'Amount']].groupby('Time').agg(['count', 'sum']).sum())

# Calculando Quantidade e Montante do Segundo Dia das transaçoes fraudulentas
d2n = list(df_work.loc[(df_work.Class == 0) & (df_work.Time >= 86400),
                  ['Time', 'Amount']].groupby('Time').agg(['count', 'sum']).sum())

# Montagem da tabela Fraude
d_f = {'Dia_1': d1f, 'Dia_2': d2f, 'Total': d1_d2f}
daf = pd.DataFrame(data=d_f, index=['Contagem', 'Montante'])
print(f'\n\033[1;30mTabela de Transações Fraudulentes\033[0;30m\n')
print(daf)

# Montagem da tabela Normal
d_n = {'Dia_1': d1n, 'Dia_2': d2n, 'Total': d1_d2n}
dan = pd.DataFrame(data=d_n, index=['Contagem', 'Montante'])
print(f'\n\033[1;30mTabela de Transações Normais\033[0;30m\n')
print(dan)
```

```
# Calculando média de operações normais por dia
media_operacoes_norm = round(((d1n[0] + d2n[0])/2),2)

# Calculando média de operações fraude por dia
media_operacoes_fraud = round(((d1f[0] + d2f[0])/2),2)

# Calculando média de Valor por Operação Normal
media_valor_ope_normal = round((d1_d2n[1]/d1_d2n[0]),2)

# Calculando média de Valor por Operação Fradulenta
media_valor_ope_fraud = round((d1_d2f[1]/d1_d2f[0]),2)

print(f'\n\033[1;30mMédia diária de Operações Normal:\033[0;30m\t\t {media_operacoes_norm: >22}')
print(f'\033[1;30mMédia de Valor por Operação Normal:\033[0;30m\t\t\t € {media_valor_ope_normal: >13}')
print(f'\033[1;30mMontante acumulado por Operação Normal em um ano:\033[0;30m\t €\
 {round((365*media_operacoes_norm*media_valor_ope_normal),2)}')

print(f'\n\033[1;30mMédia diária de Operações Fradulenta:\033[0;30m\t\t {media_operacoes_fraud: >22}')
print(f'\033[1;30mMédia de Valor por Operação Fradulenta:\033[0;30m\t\t\t € {media_valor_ope_fraud: >13}')
print(f'\033[1;30mMontante acumulado por Operação Fradulenta em um ano:\033[0;30m\t €\
 {round((365*media_operacoes_fraud*media_valor_ope_fraud),3): >12}')
```

Acima podemos ver o prejuízo causado pelas operações fraudulentas para as empresas de Cartão de Crédito. Por dia, temos **246** operações de fraude que causam também em média um prejuízo de **€ 122,21**, um número maior quando comparado a média de operações Normais **€ 88,29**.

Ao final de um ano, as instituições financeiras chegam a acumular um montante de quase **11 milhões de €** em prejuízos. Isso mostra a importância em um sistema de detecção de fraudes.

[Isso foi diretamente da Wikipedia]. A multicolinearidade ocorre quando seu modelo inclui vários fatores que estão correlacionados não apenas à sua variável de destino, mas também entre si.

Analisar a correlação entre as variáveis é muito importante para determinar qual tipo de modelo de Machine Learning poderemos usar em nosso projeto. O fato de apresentar baixa correlação entre grande parte das variáveis nos faz poder observar modelos como:

### **Dividindo o Dataset em Treino e Teste**

Criaremos um novo Dataframe `df_model` e o dividiremos em Dataset de Treino e Teste utilizando o método `train_test_split` do ScikitLearn, onde teremos 70% para Treino e 30% para Teste.

```
# Criando uma cópia do df_work para trabalhar com a criaçao de modelos de Machine Learning
df_model = df_work.copy()

# Separando os dados entre Matrix de Características e Vetor de valores targets
X = df_model.drop('Class', axis=1)
y = df.Class

# Dividindo o Dataset em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, train_size=0.70, random_state=42)
```

### **Normalizando o Conjunto de Dados**

Para normalizar o Conjunto de Dados utilizaremos o módulo `StandardScaler` do Scikit Learn. O StandardScaler é uma ferramenta comum ao trabalhar com problemas de classificação como este. Ele transforma os dados para onde há uma média de 0 e um desvio padrão de 1, padronizando assim os dados em uma distribuição normal.

```
# Instanciando método StandardScaler para a variavel scaler
scaler = StandardScaler()

# Usando o Scaler para padronizar o Datset de Treino
scaler = StandardScaler().fit(X_train)
X_train_padr = scaler.transform(X_train)

# Usando o Scaler para padronizar o Datset de Teste
X_test_padr = scaler.transform(X_test)
```

### **Balancear o Conjunto de Dados**

Ao lidar com conjuntos de dados desequilibrados, existem três técnicas comuns para equilibrar os dados:

  * Sub-Amostragem ou Redução da Classe Majoritária
  * Sobre-Amostragem ou Aumento  na Classe Minoritária
  * Combinação de Sub-Amostragem ou Redução da Classe Majoritária e Sobre-Amostragem ou Aumento da Aumento na Classe Minoritária

Para este projeto utilizaremos um tipo de técnica de sobreamostragem, chamada de método SMOTE.

>***SMOTE*** (Synthetic Minority Over-sampling Technique) é um tipo de procedimento de Sobre-Amostragem usado para corrigir os desequilíbrios nos grupos. Essa técnica cria novas instâncias de dados dos grupos minoritários, copiando as instâncias minoritárias existentes e fazendo pequenas alterações nelas. Isso torna o SMOTE excelente para amplificar sinais que já existem nos grupos minoritários, mas não criará novos sinais para esses grupos.

A técnica **SMOTE** está presente na biblioteca `imbalanced-learn`do Python o qual utilizaremos logo abaixo.

Observação: Também é importante resaltar que o balanceamento de dados devem ser feito **APÓS** a divisão dos dados em Treino e Teste pois a amostragem excessiva (no caso de Sobre-Amostragem) antes de dividir os dados pode permitir que as mesmas observações exatas estejam presentes nos conjuntos de teste e trem. Isso pode permitir que nosso modelo simplesmente memorize pontos de dados específicos e cause overfitting e má generalização dos dados de teste.


```
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
X_train_SMOTE, y_train_SMOTE = smote.fit_sample(X_train_padr, y_train)
```

#### **Naive Bayes**

Naive Bayes é um modelo de Machine Learning probabilístico usado para tarefas de classificação. O ponto crucial do classificador é baseado no teorema de Bayes.

**Teorema de Bayes**

$$P(A|B) = \frac{P(B|A)P(A)}{P(B)}\
$$

Usando o teorema de Bayes, podemos encontrar a probabilidade de A acontecer, dado que B ocorreu. Aqui, B é a evidência e A é a hipótese. A suposição feita aqui é que os preditores / recursos são independentes. Ou seja, a presença de um determinado recurso não afeta o outro. Por isso é chamado de ingênuo ou Naive em inglês.

Agora, vamos abaixo aplicar o modelo de Naive Bayes.

## **Avaliando o Desempenho de cada Modelo**

Para avaliar os modelos de Machine Learning utilizados neste projeto, optaremos por utilizar alguns métodos bastante conhecidos, com isso utilizaremos estas abreviações nas fórmulas de métrica:

* *TP* = Verdadeiro Positivo. Transações Normais que o modelo prevê corretamente.
* *TN* = Verdadeiro negativo. Transações Fraudulentas que o modelo prevê corretamente.
* *FP* = Falso Positivo. Transações Fraudulentas que o modelo prevê como Normais.
* *FN* = Falso Negativo. Transações Normais que o modelo prevê como Fraudulentas.

Vamos explicar brevemente cada uma das métricas:

1. **Acurácia**
  * A Acurácia é uma métrica para avaliar modelos de classificação. É a fração das previsões que o modelo acerta. Embora a Acurácia possa parecer uma boa métrica para medir o desempenho de um modelo, há uma grande desvantagem ao usá-lo neste conjunto de dados desequilibrado. Se eu dissesse que todas as 285.000 transações são transações normais (nenhuma fraudulenta), eu teria uma precisão de 99,8%! O problema aqui, no entanto, é a falha óbvia de todas as transações fraudulentas. Especialmente para conjuntos de dados desequilibrados, existem métricas mais eficazes para usar. Ela pode ser medida pela fórmula:

$$Acurácia = \frac{TP + TN}{TP + TN + FP + FN}\
$$

2. **Precisão**
  * A precisão responde à pergunta: ***De todas as transações previstas como fraudulentas ou nomais, qual porcentagem foi realmente fraudulenta ou normal?***
  A fómula para determinar a precisão pode ser escrita como:

$$Precisão = \frac{TP}{TP + FP}\$$   

3. **Recall**
  * A Recall responde à pergunta: ***Dentre as transações fraudulentas ou normais, qual porcentagem delas é corretamente identificada por nosso modelo?***
  A Recall é dada pela fórmula:

$$Recall = \frac{TP}{TP + FN}\$$

4. **F1-Score**
  * A F1-Score combina Recall e Precisão em uma métrica como uma média ponderada das duas. Ao contrário da Recall e da Precisão individualmente, F1-Score leva em consideração tanto falsos positivos quanto falsos negativos. Em classes desequilibradas como esta, F1 é muito mais eficaz do que a precisão para determinar o desempenho do modelo. Seu valor varia entre 0 e 1 onde 0 é o pior valor para um modelo e 1 o melhor valor.
  Podemos calcular F1-Score pela fórmula:

$$F1Score = 2*\frac{Recall * Precisão}{Recall + Precisão}\$$  

5. **Matriz de Confusão**
  * Normalmente, uma matriz de confusão é uma visualização de um modelo de classificação que mostra quão bem o modelo previu os resultados em comparação com os originais. Normalmente, os resultados previstos são armazenados em uma variável que é então convertida em uma tabela de correlação. Usando a tabela de correlação, a matriz de confusão é traçada na forma de um mapa de calor.
